# Important functions and methods for complex data types

This unit serves as a recap to summarize the key functions and methods for 
[Lists](https://docs.python.org/3/library/stdtypes.html#lists), 
[Tuples](https://docs.python.org/3/library/stdtypes.html#tuples), and 
[Dictionaries](https://docs.python.org/3/library/stdtypes.html#mapping-types-dict).

The tables below provide an overview of various methods and functions, along with 
an indication of whether they are applicable to the data types mentioned above:

| **Method**         | **Description**                           | **List** | **Tuple** | **Dictionary** |
| :----------------- | :---------------------------------------- | :------: | :-------: | :------------: |
| `.sort()`          | Sorts the object                          |    X     |           |                |
| `.append(x)`       | Appends *x* to the end of the object      |    X     |           |                |
| `.pop(i)`          | Removes item at index *i* from the object |    X     |           |       X        |
| `.insert(i, x)`    | Inserts *x* at index *i*                  |    X     |           |                |
| `.remove(x)`       | Removes the first occurrence of *x*       |    X     |           |                |
| `.count(x)`        | Counts the number of occurrences of *x*   |    X     |    X      |                |
| `.index(x)`        | Returns the index of *x*                  |    X     |    X      |                |
| `.keys()`          | Returns all `keys` of the dictionary      |          |           |       X        |
| `.values()`        | Returns all `values` of the dictionary    |          |           |       X        |
| `.items()`         | Returns all `items` of the dictionary     |          |           |       X        |



| **Function**       | **Description**                           | **List** | **Tuple** | **Dictionary** |
| :----------------- | :---------------------------------------- | :------: | :-------: | :------------: |
| `del object[item]` | Deletes *item* from *object*              |    X     |           |       X        |
| `len(object)`      | Returns the length of *object*            |    X     |    X      |       X        |
| `min(object)`      | Returns the smallest item in *object*     |    X     |    X      | [<sup id="fn1-back">1</sup>](#fn1) |
| `max(object)`      | Returns the largest item in *object*      |    X     |    X      | [<sup id="fn1-back">1</sup>](#fn1) |
| `sorted(object)`   | Returns a sorted list of the items in *object*    |    X     |    X      | [<sup id="fn1-back">1</sup>](#fn1) |

It should be evident why the first five **methods** apply only to lists — they modify 
the object, and since tuples are immutable, these methods are unavailable for them.  
Regarding dictionaries, sorting is unnecessary because values are accessed using a 
`key`, not an index. Inserting values into a dictionary uses different syntax, and 
removing the first occurrence of a value doesn't make sense because dictionaries 
can have any order, and items are accessed using keys.  
The last set of **methods** is specific to dictionaries, as lists and tuples do not 
have keys and values. Therefore, these methods work only for dictionaries.

The `del` keyword operates only on mutable data types and not on tuples. 


## Example

The methods mentioned above can be used to identify which letter occurs most often 
in a text. Note that this example is solely to demonstrate the use of these 
methods. A dictionary would provide a more efficient solution.

```python
text = "lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua. Ut enim ad minim veniam, quis nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat."
letters = "abcdefghijklmnopqrstuvwxyz"

result = []
for l in letters:
    result.append(text.count(l))

print(result)

max_count = max(result)
print(max_count)

max_index = result.index(max_count)
print(max_index)
print(letters[max_index])
```
