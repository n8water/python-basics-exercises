# What are dictionaries?

In both lists and tuples, elements are accessed using an index. This method has a
drawback: to access a specific element, you need to know its exact index, which 
can be challenging when elements are added or removed. In such cases, the indices 
of subsequent elements change, making programming cumbersome due to constant 
index tracking.

## Accessing Elements by Name

Dictionaries offer a solution to this problem. Like lists and tuples, dictionaries 
can store multiple elements. The key difference is that elements in a dictionary 
are accessed by a name rather than an index, using a key-value format.

For example, consider a telephone book. A telephone book lists numerous phone 
numbers, each associated with a name:

| Name         | Telephone Number |
| :----------- | ---------------: |
| P. McCartney |          123 456 |
| J. Lennon    |      987 654 321 |
| G. Harrison  |       11 342 555 |
| R. Starr     |       777 888 32 |

In a telephone book, you don't seek the third element; instead, you look for 
the number corresponding to a specific name, like *G. Harrison*. Essentially, you 
want the number using the name directly. Accessing an element by name is what dictionaries are good for.  Instead of "names," we call them "keys" in dictionary terminology.

## Using Dictionaries

A dictionary is a collection of key-value pairs. Dictionaries are enclosed in 
curly braces `{}`. Inside the braces, key-value pairs are separated by commas, 
and each key-value pair is of the form: `key: value`. A dictionary looks like 
this: `{key1: value1, key2: value2, ..., keyN: valueN}`. Here's an example:


````python
phone_book = {
    "P. McCartney": 123456,
    "J. Lennon": 987654321,
    "G. Harrison": 11342555,
    "R. Starr": 77788832,
}
print(phone_book)
````

You can access individual elements of `phone_book` using the key (person's name). 
To do this, place the key in square brackets `[ ]` following the dictionary's 
name.

````python
print(phone_book["G. Harrison"])
````

Access by index, however, is not possible, as shown below:

````python
print(phone_book[2])
````

Attempting to use an index with a dictionary will lead to an error, as dictionaries
do not have this index-based access. Instead, they rely on keys, and accessing a 
non-existing key results in a key error.

Here's another example demonstrating this point:


````python
print(phone_book["P. Best"])
````

### Adding key-value pairs to a dictionary

You can easily add new key-value pairs to a dictionary using the following syntax: 
`dictionary[key] = value`. In the example below, additional phone numbers are 
added to our telephone book:

````python
phone_book["Y. Ono"] = 5463333
phone_book["B. Geröllheimer"] = 9998777
print(phone_book)
````

### Replacing the value for an existing key

If you assign a new value to an already existing key in a dictionary, the value 
is overwritten. This feature is useful for updating or changing entries but be 
careful not to overwrite the wrong entries in your dictionary!

````python
phone_book["P. McCartney"] = 654321
print(phone_book)
````

### More about keys and values

In our telephone book example, all keys were strings and all values were integers. 
However, dictionaries are not limited to these data types. *Values* can be of any 
data type, including lists, tuples, and even other dictionaries. *Keys* can also 
be nearly any data type (more on the requirements for keys below). 
Within a dictionary, data 
types can be mixed, as demonstrated in the example below:


````python
stupid_dict = {
    123: "blabla",
    True: 123.456,
    "key": False,
    (123, "abc"): 1000,
    34.5: [0, 1, 1, 2, 3, 5],
}
print(stupid_dict)
````

### The key in a dictionary is immutable

In contrast to values, keys in a dictionary must be immutable.[<sup id="fn1-back">1</sup>](#fn1) Acceptable key 
types include `integers`, `strings`, and `tuples`, but not lists, since lists 
can be modified. Consider these two similar examples:

````python
new_phone_book = {("Paul", "McCartney"): 123456}
print(new_phone_book)
````

````python
new_phone_book = {["Paul", "McCartney"]: 123456}
print(new_phone_book)
````

### Deleting an element
You can delete elements from a dictionary using the `del` keyword. Review the 
example below. What happens if you run this cell twice? Why does the error occur?

````python
del phone_book["P. McCartney"]
print(phone_book)
````

### Iterating over a dictionary with a `for` loop

Just like lists and tuples, dictionaries can be iterated over using a `for` loop. 
The syntax is as follows:

````python
for name in phone_book:
    print(name, phone_book[name])
````

## Usage of functions and methods with dictionaries

Several functions and methods can be used to manage dictionaries.

### General functions and methods

| Function/Method | Return Value                                |
| --------------- | ------------------------------------------- |
| `len()`         | Number of key-value pairs in the dictionary |
| `.keys()`       | All keys of a dictionary                    |
| `.values()`     | All values of a dictionary                  |
| `.items()`      | All `key:value` pairs as tuples             |

> **NOTE:** The `.keys()` method returns a `dict_keys` data type. You can convert 
this to a list with the `list()` function.


````python
print(len(phone_book))
````

````python
print(phone_book.keys())
print(type(phone_book.keys()))
print(list(phone_book.keys()))
````

````python
print(list(phone_book.values()))
````

````python
print(phone_book.items())
print(phone_book)
del phone_book["B. Geröllheimer"]
print(phone_book)
````

## Typical applications of dictionaries

Below are two scenarios where dictionaries are particularly useful.

### Example 1: Student data

In the telephone book example, the key-value pairs were *name - telephone number*. 
In this example, a dictionary represents a student's data, which includes name, 
first-name, ID, e-mail, city, etc.

Here, the keys and values vary significantly. Each key describes a different 
attribute of the same student. The advantage is you don't need to remember the 
indices of various attributes. Unlike tuples, attributes can be added or 
removed as needed.


````python
student = {
    "name": "McCartney",
    "firstname": "Paul",
    "subject": "Music",
    "email": "paul@beatles.co.uk",
    "city": "Liverpool",
}
print(student)
print(student["name"])
print(student["city"])
````

### Example 2: Students and student IDs

When data has a unique identifier (ID), it can serve as the dictionary key. For 
example, each student can have a student ID as a key:

````python
students = {
    12345: "Paul McCartney",
    23456: "John Lennon",
    34567: "George Harrison",
    45678: "Ringo Starr",
}
print(students)
print(students[34567])
````

One of the nice features of Python is that it supports Unicode. Therefore it is possible to use emojis just like other characters in strings. In this exercise you will use this feature to build an emoji translator.

Below is a dictionary that maps English terms to Emojis (broken into multiple lines for better readability).

{
"happy": "😃",  
"heart": "😍",  
"rotfl": "🤣",  
"smile": "😊",  
"crying": "😭",  
"kiss": "😘",  
"clap": "👏",  
"grin": "😁",  
"fire": "🔥",  
"broken": "💔",  
"think": "🤔",  
"excited": "🤩",  
"boring": "🙄",  
"winking": "😉",  
"ok": "👌",  
"hug": "🤗",  
"cool": "😎",  
"angry": "😠",  
"python": "🐍"  
}    

Use this dictionary to build a program that:

    Reads a sentence from the user.
    Replaces all the words in the sentence with the corresponding Emoji.

Below is an example execution of the program:

Please enter a sentence: I'm so excited to learn python  
I'm so 🤩 to learn 🐍


````python
emoji_dict = { "happy": "😃", "heart": "😍", "rotfl": "🤣", "smile": "😊", "crying": "😭", "kiss": "😘", "clap": "👏", "grin": "😁", "fire": "🔥", "broken": "💔",
"think": "🤔", "excited": "🤩", "boring": "🙄", "winking": "😉", "ok": "👌", "hug": "🤗", "cool": "😎", "angry": "😠", "python": "🐍"} 

user_input = input("Please enter a sentence: ")

word_list = user_input.split()
dict_keys = emoji_dict.keys()

result = ""

for word in word_list:
    if word in dict_keys:
        emoji = emoji_dict[word]
        word = emoji
    result += " " + word    

result.strip()
#result = [result = result + item for item in word_list]
print(result)
````
