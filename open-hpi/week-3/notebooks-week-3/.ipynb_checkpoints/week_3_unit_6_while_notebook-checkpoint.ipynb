# The [while](https://docs.python.org/3/reference/compound_stmts.html#while) loop

You are already familiar with the `for` loop. There is another loop in Python, the
`while` loop. The `while` loop offers more flexibility, but can be a bit more
complex to use.

The `for` loop works well when you have a sequence to iterate over, like a list or
a range. However, there are scenarios where you need to repeat an operation multiple
times without a predefined sequence:
- At an ATM, the system prompts for a PIN until it is entered correctly.
- A user must enter a value that can be converted into a number.
- A user should keep answering questions until none remain. In such cases, a `while`
  loop is a better choice than using a `for` loop with "tricks".

## Syntax of the `while` loop

The simplest form of a `while` loop is structured as follows:

```python
while condition:
    statement1
    statement2
    ...
    statementN
```

The loop begins with the keyword `while`, followed by a condition and a colon, much
like an `if` statement. The subsequent block of code should be indented, similar to
other control structures. When the loop's indented code is done executing, the program
resumes execution with the non-indented code following the loop.

## How a `while` loop works

To start, the loop checks the condition specified after `while`. If the condition is
`True`, the loop's body is executed. Upon reaching the end of the loop body, the
program returns to the `while` condition and checks it again.  
The loop continues running as long as the condition remains `True`. Once the condition
becomes `False`, the loop ends and the program continues with the code following the
loop.  
If the condition is `False` the first time the `while` loop is encountered, the loop's
body is skipped entirely and the program continues immediately after the loop.

## Example 1: Using `input()` until a number is entered

Consider this problem: You need to enter a number using `input()`. For calculations,
the entered string must first be converted to a number, such as an integer using
`int()`. If the user enters non-numeric input, the conversion will cause an error.


````python
number = input("Please enter a number: ")
number = int(number)
````

To avoid this, we can use a `while` loop. The loop's condition checks if the input is
numeric using the `.isdecimal()` method. Until a valid number is entered, the loop can
keep requesting input. Once a number is entered, conversion occurs **after** the loop.

````python
# The loop runs until a suitable input is available
number = "x"
while not (number.isdecimal()):
    number = input("Please enter number: ")

number = int(number)
print(number)
````

An *imperfection* in the loop is visible above: To check the condition initially, the
variable `number` must be declared first but with a value that is not a number.
Setting `number = "x"` is solely to initialize it with something besides an integer.

Alternatively, this variant is possible:

````python
# The loop runs until a suitable input is available
number = input("Please enter number: ")
while not (number.isdecimal()):
    number = input("Please enter number: ")

number = int(number)
print(number)
````

Although the statement in line two is less "meaningless," there is now a
duplicated `input` statements. This is not the most elegant programming style. 
Question: Why can the `int(number)` conversion occur only after the loop?

## Example 2: Check PIN

In this example, we have a secret PIN. The user must enter the correct PIN to proceed
beyond the `while` loop. The program structure is similar to the previous example.


````python
secret_pin = 1234

pin = int(input("Please enter PIN: "))

while pin != secret_pin:
    print("The PIN was wrong.")
    pin = int(input("Please enter PIN: "))

print(pin, "is correct")
````

In real life, attempts at entering a PIN are usually limited. ATMs typically allow 
only three tries before blocking. How can we represent this in our loop?

We need a condition that checks both if the PIN is correct and if attempts exceed 
three. Here's how this can be implemented:

````python
secret_pin = 1234

pin = int(input("Please enter PIN: "))
attempts = 1

while (pin != secret_pin) and (attempts < 3):
    print("The PIN was wrong.")
    pin = int(input("Please enter PIN: "))
    attempts += 1

if pin == secret_pin:
    print(pin, "is correct")
else:
    print("You entered the wrong PIN three times, your card will be confiscated.")
````

This program also has *imperfections*. A second variable `attempts` is required, 
initialized before the loop and incremented inside it. After the loop, it's not 
directly clear why it ended — was it a correct PIN or exceeded attempts? An `if` 
statement is needed to clarify this.

## Example 3: Simple counter

In the previous example, a counter (`attempts`) was used, illustrating the basic act 
of incrementing variables. Here's how you can count from 1 to 10 with a `while` loop:


````python
# Simple counter
# The following program is to count from 1 - 10
i = 1
while i <= 10:
    print(i)
    i += 1
````

This solution appears straightforward. However, subtle details can 
alter program behavior:
- Should `i` start at 1 or 0?
- Should the condition be `i <= 10` or `i < 10`?
- Should the comparison value be 10 or 11?
- Should the increment (`i += 1`) occur before or after `print()` in the loop body?

Each small variation changes how the program behaves. Therefore, when applying `while`
loops, pay attention to boundary conditions. Experiment by modifying the above program
with these questions in mind and predict its output.

always remains `True`. For instance, in the previous example, if the increment is 
forgotten, the loop becomes infinite.

````python
# The following program is to count from 1 - 10
i = 1
while i <= 10:
    print(i)
    # Because of the commented out (forgotten) increment, the loop runs endlessly.
    # i += 1
````

Using a `for` loop is simpler and more robust as it automatically checks boundaries. 
Use a `for` loop when possible; it is simpler and thus reduces potential errors.

# Example 4: Guessing a random number

A `while` loop is well-suited for implementing a number guessing game.
In this game, a random number between 1 and 100 is generated. Unlike the previous PIN example, the secret number 
is unknown to the user. After each guess, the program provides a hint indicating whether the secret number is higher 
or lower than the user's guess. This process continues until the user correctly guesses the number.


````python
import random

secret_number = random.randint(1, 100)

guessed_number = int(input("Please guess a number: "))

while guessed_number != secret_number:
    if guessed_number < secret_number:
        print("The number", guessed_number, "was too small.")
    else:
        print("The number", guessed_number, "was too big.")

    guessed_number = int(input("Please guess a number: "))

print("Correct!", guessed_number, "was the number you were looking for.")
````

# Exiting the loop with a `break`

A while loop can be exited early using the break statement. When break is executed, the loop stops immediately — even if the loop condition is still true. After a break, the condition is no longer checked. Therefore, break is typically used inside an if statement within the loop.

Using break can simplify certain programming tasks. For example, suppose we want to enter data for multiple students, where each student has a matriculation number, last name, and first name. Since we don’t know in advance how many students will be entered, a while loop is a good choice.

We want the loop to stop when the user presses the return key without entering a matriculation number. This can be implemented cleanly using break, as shown below:

````python
list_of_students = []

while True:
    matnr = input("Enter matriculation number: ")
    if matnr == "":
        break
    name = input("Enter name: ")
    firstname = input("Enter first name: ")
    list_of_students.append((matnr, name, firstname))

print(list_of_students)
````

`while True:` sets up an infinite loop since `True` is always true. The loop only exits 
via `break`. After the first `input()`, a condition checks whether to exit. If so, 
the `break` ends the loop and prevents further `input()` calls.
