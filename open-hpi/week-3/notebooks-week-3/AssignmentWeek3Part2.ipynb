# Ceasar cipher
It' allowed to handle all transform input to lower case only. Do not alter whitespace and special characters



```python
""" Ceasar ciper assignment """
def encrypt(message, shift):
    """ Returns encrypted message where the letters are shifted 'shift' positions """
    message = message.lower()
    alphabet = "abcdefghijklmnopqrstuvwxyz"

    result = ""
    
    for letter in message:
        if letter in alphabet:
            index = alphabet.find(letter)
            new_index = index + shift
            if new_index > 25:
                new_index -= 26
            elif new_index < 0:
                new_index += 26
            result = result + alphabet[new_index]
            continue
        result = result + letter

    return result

number_to_shift = "x"

while not number_to_shift.isdecimal():
    number_to_shift = input("Please enter the number of places to shift: ")

number = int(number_to_shift)
user_message = input("Please enter a senctence: ")
encrypted_message = encrypt(user_message, number)
print("The encrypted sentence is:", encrypted_message)

            
    
```
