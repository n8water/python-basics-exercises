# When to use lists, tuples and dictionaries

So far, we have introduced three complex data types: `list`, `tuple`, and `dictionary`. 
In principle, these data types are interchangeable.[<sup id="fn1-back">1</sup>](#fn1)

For instance, a list can be simulated with a dictionary by using the index of a 
list item as the dictionary key. However, this approach can complicate certain 
operations. For example, if you delete an item from a list, all subsequent indices 
decrease by one. Simulating this behavior with a dictionary requires complex 
operations. Similarly, a tuple can also be mimicked with a list, but using a list 
sacrifices the immutability that is intrinsic to a tuple. So, why not opt for 
lists initially if they offer more possibilities? What advantages do different 
complex data types provide, and when should each type be used?

The following suggestions are based on personal experiences and programming styles. 
These recommendations should not be taken as the definitive truth but merely as 
guidance. These recommendations may not be shared universally by other programmers,
and that's perfectly fine.[<sup id="fn2-back">2</sup>](#fn2)

## Lists should be used for many objects

A list can contain items of different data types, such as `[23, "abc", True, ("a", "b", "c")]`. 
However, handling lists with varying data types can be challenging. If all elements 
share the same data type, managing the list becomes easier. Knowing the data type 
of list elements allows for consistent use of functions and methods. For example, 
consider the phone number list from a previous notebook. Iterating through this 
list ensures you're processing phone numbers, typically integers. Limiting lists 
to a single data type can simplify programming.

In the example below, the list `numbers` contains only integers. The `for` loop 
takes advantage of this knowledge and divides each value by 2 using integer division 
`//`. If the list contained other types, like strings or lists, this operation 
would result in an error.


````python
numbers = [2, 4, 8, 16, 32, 64]
for i in numbers:
    print(i // 2)
````

## Tuples should be used for complex objects

As mentioned in the first notebook of this week, tuples are best suited for complex 
objects requiring multiple attributes. Take students as an example: each student 
might be described by a name, first name, email, student ID, course of study, 
address, etc. These attributes are consistent across all students, making tuples 
ideal for representing each student.

If your program manages multiple students, you can store each student tuple in a 
list. This combination of lists and tuples—a list of tuples—is effective for 
handling multiple complex objects of the same type.

The list-of-tuples structure works well when accessing data from CSV files (CSV 
stands for comma-separated values), a common data format.

````python
student_1 = ("Dylan", "Bob", 334455, "Philosophy")
student_2 = ("Cobain", "Kurt", 987654, "Mechanical Engineering")
student_3 = ("Winehouse", "Amy", 123321, "Medicine")
list_of_students = [student_1, student_2, student_3]

for s in list_of_students:
    print(s[3])
````

## If there is an ID, use a dictionary instead of a list

When objects have a unique ID, a dictionary may be more suitable. This allows you 
to access objects directly via their IDs rather than searching through them.

Consider the example of students: each student has a unique student ID. Operations 
with students often rely on their IDs rather than names to avoid confusion—especially 
among students with identical names. Unlike names, student IDs are unique. 
Therefore, for student data, a dictionary of tuples can be advantageous.

A dictionary of tuples is also a good option for data retrieved from relational 
databases. These databases typically consist of records with consistent structures 
and unique IDs. The ID can serve as the dictionary key, and the rest of the record, 
in tuple form, becomes the dictionary value.


````python
student_1 = ("Dylan", "Bob", "Philosophy")
student_2 = ("Cobain", "Kurt", "Mechanical Engineering")
student_3 = ("Winehouse", "Amy", "Medicine")

dict_of_students = {}
dict_of_students[334455] = student_1
dict_of_students[987654] = student_2
dict_of_students[123321] = student_3

for student_ID in dict_of_students:
    print(student_ID, dict_of_students[student_ID][2])
````

## If complex objects differ in their attributes, use a dictionary instead of the tuple

Unlike tuples, dictionaries provide greater flexibility in describing objects. 
Consider again the example of students: students from different departments may 
take different classes and receive marks in distinct modules. In such scenarios, 
dictionaries can replace tuples. If no ID is present, a list of dictionaries is 
viable; if an ID is available, a dictionary of dictionaries is feasible.

Even more complex combinations are possible. For instance, students typically share 
attributes like name, first name, email, etc., making tuples appropriate. However, 
students enroll in different courses, and their modules and achievements vary. 
Here, a dictionary for these attributes is preferable. A combination of tuples 
and dictionaries is possible: student data is stored in a tuple, while one tuple 
element, containing a student's marks, is a dictionary. You end up with a list 
of tuples, where one element of the tuple is a dictionary. Complex, but effective.

There are many options, but choosing a well-suited data structure often makes 
your life as a programmer easier. So, carefully consider your data structure choices.

````python
s1 = ("Dylan", "Bob", 334455, "Philosophy", {"Logic": "A", "Ethics": "B"})
s2 = ("Cobain", "Kurt", 987654, "Mechanical Engineering", {"Math": "B"})
s3 = ("Winehouse", "Amy", 123321, "Medicine", {"Math": "B", "Chemistry": "C"})

l_of_students = [s1, s2, s3]
for s in l_of_students:
    print(s)
````
