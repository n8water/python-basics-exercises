# Safe dictionary access

When you try to access a key that does not exist in a dictionary, you will
encounter a `KeyError`. This is demonstrated in the following example:

```python
dict = {1234: "Peter", 2345: "Jane"}
print(dict[1234])
print(dict[3456])
```

## Safe access to a dictionary using `in`

To prevent your program from being interrupted by errors, it is a good idea to
catch these potential errors. Using the keyword `in`, you can check if a key
exists in the dictionary, and thus avoid the error entirely. The expression
`key in dict` will return `True` if the key exists in the dictionary, and
`False` otherwise. You can use this within an `if` statement to handle the
situation gracefully.

```python
dict_of_students = {
    477264: ("Potter", "Harry", "harry@hogwarts.wiz", "Defence Against the Dark Arts"),
    490134: ("Weasley", "Ron", "ron@hogwarts.wiz", "Care of Magical Creatures"),
    471617: ("Granger", "Hermione", "hermione@hogwarts.wiz", "Alchemy"),
    432646: ("Creevey", "Colin", "colin@hogwarts.wiz", "Music"),
    481989: ("Finnigan", "Seamus", "seamus@hogwarts.wiz", "Ancient Studies"),
    488962: ("Abbott", "Hannah", "hannah@hogwarts.wiz", "Apparition"),
    482103: ("Parkinson", "Pansy", "pansy@hogwarts.wiz", "Dark Arts"),
    492010: ("Malfoy", "Draco", "draco@hogwarts.wiz", "Defence Against the Dark Arts"),
    447924: ("Thomas", "Dean", "dean.thomas@hogwarts.wiz", "Divination"),
}

matrnr = int(input("Please enter matriculation number: "))
if matrnr in dict_of_students:
    print("The student you are looking for is:", dict_of_students[matrnr])
else:
    print("A student with this matriculation number does not exist")
```

## Alternative: Using [dict.get()](https://docs.python.org/3/library/stdtypes.html#dict.get) to access values safely

Another way to avoid errors is by using the `get()` method to retrieve the
value associated with a given key. This method can take two arguments:

`dict.get(key, default)`

The first argument is the `key` you want to retrieve the value for (such as the
matriculation number in the above example). The second, **optional** argument,
`default`, specifies the return value if the `key` is not found in the
dictionary. If you do not provide the `default` argument, the method will 
return `None` by default, and no error will be raised, so your program will 
continue to run smoothly. Below is an example of how to implement this 
technique:

```python
input_num = int(input("Please enter matriculation number:"))
student = dict_of_students.get(
    input_num, "no matching student found for this matriculation number"
)

print(
    "For the matriculation number",
    input_num,
    "the dictionary returned the following result:",
    student,
)
```
