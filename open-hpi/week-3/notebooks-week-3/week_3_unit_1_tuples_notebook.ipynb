# What are [tuples](https://docs.python.org/3/library/stdtypes.html?#tuples)?

## Definition

A `tuple` is a complex data type, similar to lists, but with a key difference:
tuples are immutable, while lists are mutable. This means that once a tuple is
created, its elements cannot be changed.

A tuple consists of multiple elements enclosed in parentheses: `()`. The
elements are separated by commas. Just like with lists and simple data types,
tuples can be assigned to variables. Additionally, tuples can store other tuples
or lists as elements.


```python
address = (52066, "Aachen", "Eupener Str. 70", "0241-6009-12345")
print(address)
student = ("Peter", "Meier", 123456, "SAP Basics", "pm12345s@university.edu", address)
print(student)
tup1 = (12312, "absbsb", [1, 2, 3, 4], ("a", "b", "c"), "end")
print(tup1)
```

## Tuples cannot be modified

The key difference between tuples and lists is that lists can be modified, while
tuples cannot — they are immutable. This means you cannot add or remove elements
in a tuple or change the value of an existing element. While you can assign a
new tuple to a variable, there are no methods like `tuple.append()` or
`tuple.remove()` to alter tuples directly.


```python
tup = (1, 2, "a")
print(tup[2])
tup[2] = "b"
```

```python
tup = (1, 2, 3)
print(tup)
tup = ("a", "b", "c", "d")
print(tup)
```

## What are tuples used for?

Tuples are quite similar to lists. In this unit, we will not delve deeply into
data modeling; that will be covered in Unit 3. However, here is a simple rule of
thumb for choosing between lists and tuples:

- Use lists for collections of similar items.
- Use tuples for items with many attributes.

Consider the following examples where tuples are appropriate:

- `address = (zip_code, city, street, house_number)`
- `position = (x_coordinate, y_coordinate)`
- `date = (year, month, day)`

## Basic operations on tuples
### Creating a Tuple Interactively

You can use the `input()` function to read values and assemble them into a
tuple.


```python
name = input("Please enter name: ")
first_name = input("Please enter first name: ")
phone = input("Please enter phone number: ")
age = input("Please enter age (integer): ")

employee = (name, first_name, phone, age)

print(employee)
```

### Using an index with tuples

Just like lists, tuples use an index that is represented with square brackets
`[]`. The index starts at 0. You can also use negative indices, similar to
lists. 

**Important:** Even though you use square brackets to access elements,
it remains a tuple, not a list.


```python
address = (52066, "Aachen", "Eupener Str. 70", "0241-6009-12345")

student = (
    "Peter",
    "Parker",
    123456,
    "Python for Beginners",
    "pp12345s@university.edu",
    address,
)

print(address[0])
print(student[1])
print(student[5])
print(student[5][2])
print(student[-1])
print(address[-3])
print(address[2])
```

### Once again: Tuples are immutable

Attempting to change a single element of a tuple by using its index will result
in an error.

```python
address = (52066, "Aachen", "Eupener Str. 70", "0241-6009-12345")
address[2] = "Goethestraße 1"
```

### Slicing operator and functions & methods

The slicing operator, familiar from lists, also works with tuples to create a
sub-tuple. You can use indices to access not only a single element, but an
entire range. Some functions and methods work with tuples as well, as shown in
the table below:

| Function / Method | Return Value                                                                             |
| ----------------- | ---------------------------------------------------------------------------------------- |
| `len(tuple)`      | Number of elements in the tuple                                                          |
| `tuple.count(x)`  | Number of times element *x* appears in the tuple                                         |
| `tuple.index(x)`  | Index of the first occurrence of *x*. If *x* does not exist, an error will occur          |


```python
numbers = (1, 2, "trois", "four", "V", 6)
print(numbers[2:4])
print(len(numbers))
print(numbers.count(1))
print(numbers.index("V"))
```

### Looping through tuples

Similar to lists, a `for` loop can be used to access each element of a tuple.

```python
address = (52066, "Aachen", "Eupener Str. 70", "0241-6009-12345")

for address_part in address:
    print(address_part)
```

### Conversion of tuples

There are functions, like `int()`, to convert data types — such as from a string
to an integer. Similar conversions exist between lists and tuples: `list()`
converts the argument into a list, and `tuple()` converts it into a tuple.


```python
l = [1, 2, "a", 2.3]
t = tuple(l)
print(t)
l = list(t)
print(l)
```
